### GC Content

```python
sequence = ""

with open("SARS-CoV-ZJ0301.fasta") as myfile:
    line = myfile.readline()
    line = myfile.readline()
    
    while line!="":
        sequence += line
        line = myfile.readline()
        
g_count = sequence.count("G")
c_count = sequence.count("C")

gc_content_SARS = (g_count+c_count)/len(sequence)*100
print("GC content of SARS coronavirus ZJ0301:", gc_content_SARS)
```

```python
from sklearn.decomposition import PCA
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
np.random.seed(10)

plt.rcParams.update({'font.size': 16})

profiles = []

with open("300_trimers.txt") as file:
    for line in file.readlines():
        f_list = [float(i) for i in line.split(" ") if i.strip()]
        profiles.append(f_list)

label_vec = ["SARS-CoV"]*100
label_vec = label_vec + ["bat-SL-CoV"]*100
label_vec = label_vec + ["COVID-19"]*100


DATA=np.array(profiles)

pca = PCA(n_components=2)
embedded = pca.fit_transform(DATA)

fig = plt.figure(figsize=(8, 6))
sns.scatterplot(embedded[:,0], embedded[:,1], hue=label_vec, alpha=1, s=100, palette=['blue','red','green']).plot()   
plt.savefig('pca.png', format="png", dpi = 300, bbox_inches='tight')
```

```python
from sklearn.manifold import TSNE

embedded = TSNE(n_components=2, init='pca').fit_transform(DATA)

fig = plt.figure(figsize=(8, 6))
sns.scatterplot(embedded[:,0], embedded[:,1], hue=label_vec, alpha=1, s=100, palette=['blue','red','green']).plot()   
plt.savefig('tsne.png', format="png", dpi = 300, bbox_inches='tight')
```
