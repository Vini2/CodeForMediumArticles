## 2D plots of reads from Pseudomonas aeruginosa and Staphylococcus aureus

```python
from sklearn.decomposition import PCA
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
np.random.seed(10)

plt.rcParams.update({'font.size': 16})

profiles = []

with open("reads.txt") as file:
    for line in file.readlines():
        f_list = [float(i) for i in line.split(" ") if i.strip()]
        profiles.append(f_list)

label_vec = ["Pseudomonas aeruginosa"]*100
label_vec = label_vec + ["Staphylococcus aureus"]*100


DATA=np.array(profiles)

pca = PCA(n_components=2)
embedded = pca.fit_transform(DATA)

fig = plt.figure(figsize=(10, 5))
sns.scatterplot(embedded[:,0], embedded[:,1], hue=label_vec, alpha=1, s=100, palette=['blue','red']).plot()   
plt.savefig('pca_pa_sa.png', format="png", dpi = 300, bbox_inches='tight')
```

```python
from sklearn.manifold import TSNE

embedded = TSNE(n_components=2, init='pca').fit_transform(DATA)

fig = plt.figure(figsize=(10, 5))
sns.scatterplot(embedded[:,0], embedded[:,1], hue=label_vec, alpha=1, s=100, palette=['blue','red']).plot()   
plt.savefig('tsne_pa_sa.png', format="png", dpi = 300, bbox_inches='tight')
```

## 2D plots of reads from Enterococcus faecalis and Listeria monocytogenes
